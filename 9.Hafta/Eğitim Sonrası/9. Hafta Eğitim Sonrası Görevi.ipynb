Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python

import random

class kumanda():
    def __iniit__(self,tv="kapalı",ses=0,kanallar=["trt","show","now","atv"],kanal="trt"):
        self.tv=tv
        self.se=ses
        self.kanallar=kanallar
        self.kanal=kanal
    def tvAc(self):
        if self.tv=="açık":
            print("acık")
        else:
            self.tv=="kapalı"
            print("açılıyor")
    def tvKapa(self):
        if self.tv=="kapalı"
            print("kapalı")
        else:
            self.tv="kapalı"
            print("kapatılıyor")
    def kanalEkle(self,yeniKanal):
        self.kanallar.append(yeniKanal)
        print("kanal eklendi")
    def rastgele(self):
        rastgele=random.randint(0,len(self.kanallar))
        self.kanal=self.kanallar[rastgele]
        print(self.kanal)
    def kanalSayi(self):
        return len(self.kanallar)
       
kumanda=kumanda()

print("""1. Tv Aç
2. Tv Kapat
3. Kanal Ekle
4. Kanal Sayısını Öğrenme
5. Rastgele Kanala Geçme
      Çıkmak içi 'q'ya bas
""")
while True:
    islem=input("işlem no:")

    if islem=="q":
        break
    elif islem=="1":
        kumanda.tvAc()
    elif islem=="2":
        kumanda.tvKapa()
    elif islem=="3":
        kumanda.kanalEkle()
    elif islem =="4":
        kumanda.kanalSayi()
    elif islem=="5":
        kumanda.rastgele()
        











```
